```python
from bs4 import BeautifulSoup
import pandas as pd
from datetime import datetime as dt
from selenium import webdriver
import re
```

```python
from webdriver_manager.chrome import ChromeDriverManager
```

```python
option = webdriver.ChromeOptions()
option.add_argument('--headless')
option.add_argument('--no-sandbox')
option.add_argument('--disable-dev-shm-usage')
# driver = webdriver.Chrome('chromedriver', options = option)
driver = webdriver.Chrome(ChromeDriverManager(version='95.0.4638.69').install(), options = option)
```

```python
url='https://corona.samarindakota.go.id/v2/statistik/global'
driver.get(url)
```

```python
content = driver.page_source
soup = BeautifulSoup(content,'lxml')
div = soup.findAll("table",class_="table table-striped table-bordered")
data=[i.get_text() for i in div]
data1 = []
for i in range(len(data)):
     data1.append(re.findall(r'\d+.+', data[i])) #0 Dalam Perawatan 1 Discarded 2 Kasus Terkonfirmasi 3 Meninggal 4 Probable 5 Sembuh
```

```python
dalam_perawatan = []
discarded = []
kasus_terkonfirmasi = []
meninggal = []
probable = []
sembuh = []

j = 0
for i in range(len(data1[0])):
    if j == 0 :
        dalam_perawatan.append(data1[0][i])
        discarded.append(data1[1][i])
        kasus_terkonfirmasi.append(data1[2][i])
        meninggal.append(data1[3][i])
        probable.append(data1[4][i])
        sembuh.append(data1[5][i])
    elif (j % 2 == 0):
        dalam_perawatan.append(data1[0][i])
        discarded.append(data1[1][i])
        kasus_terkonfirmasi.append(data1[2][i])
        meninggal.append(data1[3][i])
        probable.append(data1[4][i])
        sembuh.append(data1[5][i])
    j= j+1
```

```python
tanggal = []
j = 0
for i in range(len(data1[0])):
    if j == 1 :
        tanggal.append(data1[0][i])
    elif (j % 2 == 1):
        tanggal.append(data1[0][i])
    j= j+1
```

```python
tgl = []
date_update = []
bulan = ['januari', 'februari', 'maret', 'april', 'mei', 'juni','juli','agustus', 'september', 'oktober','november','desember']

for i in range(len(tanggal)):
    tgl.append(tanggal[i].split(" "))

for i in range(len(tgl)):
    bln = bulan.index(tgl[i][1].lower())+1
    date_ = tgl[i][2]+'-'+str(bln)+'-'+tgl[i][0]
    date_update.append(dt.strptime(date_,'%Y-%m-%d').strftime('%Y-%m-%d'))

now = dt.now().strftime('%Y-%m-%d')
```

```python
scrape_date = []
date_update2 = []
provinsi = []
kabkot = []
kecamatan = []
kelurahan= []
total_odp = []
total_pdp = []
total_positif = []
positif_sembuh = []
positif_dirawat = []
positif_isolasi = []
positif_meninggal = []
types = []
source_link = []
user_pic = []

for i in range(len(date_update)):
    total_pdp.append(probable[i])
    total_odp.append(discarded[i])
    positif_dirawat.append(dalam_perawatan[i])
    total_positif.append(kasus_terkonfirmasi[i])
    positif_meninggal.append(meninggal[i])
    positif_sembuh.append(sembuh[i])
    positif_isolasi.append(0)
    scrape_date.append(now)
    date_update2.append(date_update[i])
    provinsi.append('Kalimantan Timur')
    kabkot.append('Kota Samarinda')
    kecamatan.append('')
    kelurahan.append('')
    types.append('Kabkot')
    source_link.append(url)
    user_pic.append('Tubagus Ibrohim')
```

```python
#create table
df = pd.DataFrame()
df['scrape_date'] = scrape_date
df['date_update'] = date_update2
df['provinsi']= provinsi
df['kabkot'] = kabkot
df['kecamatan'] = kecamatan
df['kelurahan'] = kelurahan
df['total_odp'] = total_odp
df['total_pdp'] = total_pdp
df['total_positif'] = total_positif
df['positif_sembuh'] = positif_sembuh
df['positif_dirawat'] = positif_dirawat
df['positif_isolasi'] = positif_isolasi
df['positif_meninggal'] = positif_meninggal
df['types'] = types
df['source_link'] = source_link
df['user_pic'] = user_pic
 #Output
df
```
